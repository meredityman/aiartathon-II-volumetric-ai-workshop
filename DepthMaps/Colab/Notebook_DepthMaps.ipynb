```python
import numpy as np

import tensorflow as tf
import tensorflow_hub as hub
import urllib
from pathlib import Path
from PIL import Image

model = hub.load("https://tfhub.dev/intel/midas/v2_1_small/1", tags=['serve'])
```

```python

dir = Path("/content/drive/MyDrive/SHARED/AI Artathon 2/VideoFrames")
outputdir = Path("/content/drive/MyDrive/SHARED/AI Artathon 2/DepthFrames")

for i, path in enumerate(dir.glob("*.jpg")):
  image = np.array(Image.open(path), dtype=np.float32)
  image = image / 255.0


  input = tf.image.resize(image, [256,256], method='bicubic', preserve_aspect_ratio=False)
  input = tf.transpose(input, [2, 0, 1])
  input = tf.expand_dims(input, axis=0)

  print(i, path)
  print(input.shape)

  tensor = tf.convert_to_tensor(input, dtype=tf.float32)
  output = model.signatures['serving_default'](tensor)

  output = output['default'].numpy().reshape([256, 256])
  depth_min = output.min()
  depth_max = output.max()
  output = (255 * (output - depth_min) / (depth_max - depth_min)).astype("uint8")


  depthmap = Image.fromarray(output)
  depthmap.save(Path(outputdir, path.stem + ".png"))
```
